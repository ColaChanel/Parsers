```python
import requests
```

Write link to English Premier League stats

```python
standing_url = "https://fbref.com/en/comps/9/Premier-League-Stats"
```

This will send a request to the server and download the html for this page

```python
data = requests.get(standing_url)
```

```python
data.text
```

### parsing html links with BeautifulSoup

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(data.text)
```

to select any table elements in the page that has the class stats_table

```python
standing_table = soup.select("table.stats_table")[0]
```

```python
standing_table
```

The next step is to find all of the a tags

```python
links = standing_table.find_all("a")
```

```python
links
```

получаем все что лежит внутри блоков a, вытаскиваем ссылки

```python
links = [l.get('href') for l in links]
```

```python
links
```

получаем ссылки только на команды

```python
links = [l for l in links if '/squads/' in l]
```

```python
links
```

```python
team_urls = [f"https://fbref.com{l}" for l in links]
```

```python
team_urls
```
